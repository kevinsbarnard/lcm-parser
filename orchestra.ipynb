```python
with open('orchestra.pkl', 'rb') as f:
    orchestra = pickle.load(f)
```

```python
print(len(orchestra.get_signal('tracks')._elements))
```

```python
print(orchestra.start)
print(orchestra.end)
```

```python
tracks_orchestra = orchestra.get_signal('tracks')

longest_track_id = max(tracks_orchestra._elements.keys(), key=lambda k: tracks_orchestra._elements[k].end - tracks_orchestra._elements[k].start)
print('Longest track:', longest_track_id)
```

```python
track_orchestra = tracks_orchestra.get_signal(longest_track_id)

orchestra[track_orchestra.start]
```

```python
for dt, score in track_orchestra.get_signal('score').values:
    print(orchestra.get_signal('supervisor').get_signal('state_number')[dt], score)
```
